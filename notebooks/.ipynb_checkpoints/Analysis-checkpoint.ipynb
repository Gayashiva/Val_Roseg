```python
# -*- coding: utf-8 -*-
from __future__ import print_function
from __future__ import division
import datetime

%load_ext autoreload
%autoreload 2

%matplotlib inline
```

```python
import pandas as pd
import numpy as np
from datetime import timedelta,date, time
from matplotlib import pyplot as plt
```

### Loading weather data marked with height and water input information

```python
df = pd.read_csv('../data/interim/roseg_measurements.csv', parse_dates=['time'])
del df['doy']
df.head()
```

Calculating Growth Rate which is the rate of change in height(cm/hr)

```python
df['delta_time'] = (df['time']-df['time'].shift()).fillna(0)
df['delta_height'] = (df['height']-df['height'].shift()).fillna(0)
df['growth_rate'] = (df['delta_height']*100)/(df['delta_time'].apply(lambda x: x.total_seconds()/3600))
```

### Loading all weather records 

```python
dfb = pd.read_csv('../data/interim/roseg_data.csv', parse_dates=['time'])
dfb['lw']=dfb['lwout']-dfb['lwin']
dfb['sw']=dfb['swout']-dfb['swin']
#dfb['energy']=dfb['sw']+dfb['lw']
dfb.set_index('time', inplace=True)
dfb.head()
```

```python
print('Weather Data contains %d records' % len(dfb))
```

```python
df = df[['time', 'temp', 'height', 'growth_rate','water','lw','sw']].copy()
```

### Defining additional features
Just correlating the data to the Temperature, wind speed and humidity might not capture the variation in growth rate. Adding average temperature, humidity and Wind Speed values over the past 24 hours might be useful.

```python
df['temp_24h_avg'] = df['time'].apply(lambda x: dfb[(dfb.index<=x) & (dfb.index>=(x-timedelta(hours=24)))]['temp'].mean())
df['temp_24h_count'] = df['time'].apply(lambda x: dfb[(dfb.index<=x) & (dfb.index>=(x-timedelta(hours=24)))]['temp'].count())

df['rh_24h_avg'] = df['time'].apply(lambda x: dfb[(dfb.index<=x) & (dfb.index>=(x-timedelta(hours=24)))]['rh'].mean())
df['rh_24h_count'] = df['time'].apply(lambda x: dfb[(dfb.index<=x) & (dfb.index>=(x-timedelta(hours=24)))]['rh'].count())

df['ws_24h_avg'] = df['time'].apply(lambda x: dfb[(dfb.index<=x) & (dfb.index>=(x-timedelta(hours=24)))]['ws'].mean())
df['ws_24h_count'] = df['time'].apply(lambda x: dfb[(dfb.index<=x) & (dfb.index>=(x-timedelta(hours=24)))]['ws'].count())

df['lw_24h_avg'] = df['time'].apply(lambda x: dfb[(dfb.index<=x) & (dfb.index>=(x-timedelta(hours=24)))]['lw'].mean())
df['lw_24h_count'] = df['time'].apply(lambda x: dfb[(dfb.index<=x) & (dfb.index>=(x-timedelta(hours=24)))]['lw'].count())

df['sw_24h_avg'] = df['time'].apply(lambda x: dfb[(dfb.index<=x) & (dfb.index>=(x-timedelta(hours=24)))]['sw'].mean())
df['sw_24h_count'] = df['time'].apply(lambda x: dfb[(dfb.index<=x) & (dfb.index>=(x-timedelta(hours=24)))]['sw'].count())
```

Filtering records for which averages are calculated over less than 5 values over the past 12 hours

```python
df = df[(df[[col for col in df.columns if col.endswith('count')]]>=5).all(axis='columns')]
df = df[[col for col in df.columns if not col.endswith('count')]]
```

```python
df['type'] = df['time'].apply(lambda x: 'AM' if x.time()<datetime.time(12) else 'PM')
```

### Final Cleaning

```python
df.dropna(inplace=True)
df.head()
```

```python
df[(df['growth_rate'])>20]
```

On inspecting the corresponding images, we find that growth rate is high on 17th November as the fountain pipe was extended then whereas on 19th November it seems the fountain water discharge was unusually large.

```python
df[(df['growth_rate'])<-20]
```

All the corresponding images show fragmentation and breakup of ice at the tip. We keep these values as this is how freeze and melt works at the peak. To avoid abrupt change in values in the future the pixel area of ice should be recorded instead.

```python
df=df[(df['growth_rate'])<20]
df=df[(df['growth_rate'])>-20]
```

### Correlation/linear regression plots
With all the data loaded and features defined, lets plot all the features and estimate their linear relationship to growth rate

```python
var_dict = {'temp': 'Instantaneous temperature',
            'temp_24h_avg': 'Average temperature over previous 24 hours [C]',
            'ws_24h_avg': 'Average Wind Speed over previous 24 hours',
            'rh_24h_avg': 'Average relative humidity over previous 24 hours [%]',
            'lw_24h_avg': 'Average Longwave radiation',
           'sw_24h_avg': 'Average Shortwave radiation'}
for var in var_dict:
    fig=plt.figure(figsize=(12,8))
    ax=fig.add_subplot(1,1,1)
    for mtype in [('AM', 'Night', 'b'), ('PM', 'Day', 'r')]:
        dftmp = df
        dftmp.plot(x=var, y='growth_rate', kind='scatter', ax=ax, label=mtype[1], color='r')
        plt.plot(dftmp[var], np.poly1d(np.polyfit(dftmp[var], dftmp['growth_rate'], 1))(dftmp[var]),
                 color='b')
    plt.grid()
    ax.set_xlabel(var_dict[var])
    ax.set_ylabel('Growth rate [cm/hour]')
    plt.show();
```

As most of the freezing happens in the night and the melt in the day. Let's classify data points as night and day points

```python
for var in var_dict:
    fig=plt.figure(figsize=(12,8))
    ax=fig.add_subplot(1,1,1)
    for mtype in [('AM', 'Night', 'b'), ('PM', 'Day', 'r')]:
        dftmp = df[(df['type']==mtype[0])]
        dftmp.plot(x=var, y='growth_rate', kind='scatter', ax=ax, label=mtype[1], color=mtype[2])
        plt.plot(dftmp[var], np.poly1d(np.polyfit(dftmp[var], dftmp['growth_rate'], 1))(dftmp[var]),
                 color=mtype[2])
    plt.grid()
    ax.set_xlabel(var_dict[var])
    ax.set_ylabel('Growth rate [cm/hour]')
    plt.show();
```

Indeed there is a marked change in behoviour during the night and day periods. But there were also lot of periods when the fountain was not functioning. Lets remove these data points next.

```python
for var in var_dict:
    fig=plt.figure(figsize=(12,8))
    ax=fig.add_subplot(1,1,1)
    for mtype in [('AM', 'Night', 'b'), ('PM', 'Day', 'r')]:
        dftmp = df[(df['type']==mtype[0]) & (df['water'])==1]
        dftmp.plot(x=var, y='growth_rate', kind='scatter', ax=ax, label=mtype[1], color=mtype[2])
        plt.plot(dftmp[var], np.poly1d(np.polyfit(dftmp[var], dftmp['growth_rate'], 1))(dftmp[var]),
                 color=mtype[2])
    plt.grid()
    ax.set_xlabel(var_dict[var])
    ax.set_ylabel('Growth rate [cm/hour]')
    plt.show();
```

Seems like most of the growth rate records below 10 have dissapeared and so were recorded when the fountain was not functioning. So most growth rates below 10 were due to natural melt and precipitation.

### Multivariate linear regression
We want to know how the growth rate depends on the predictors we measured, and to what extent a linear combination of these predictors alone can explain the observed growth rate.

We will use elastic net regularisation, which has the advantage of doing feature selection and accounting for collinearity between predictors. This allows us to add new features to the model selection workflow and select the relevant ones.

```python
print('%d morning measurements' % len(df[df['type']=='AM']))
print('%d evening measurements' % len(df[df['type']=='PM']))
```

We will model separately the growth and the melt process, as they have different dynamics and depend on different parameters. In addition, our data set is so small that it won't make sense to do cross-validation, so there's nothing wrong with scoring (testing) on the training set.

```python
from sklearn.linear_model import ElasticNet
model = ElasticNet()
```

Lets fit all features to growth rate first

```python
model = ElasticNet()
X = df[[col for col in df.columns if (col.startswith('temp')
                                                           or col.startswith('rh')
                                                           or col.startswith('ws')
                                      or col.startswith('lw')
                                      or col.startswith('sw')
                                                          )]]
y = df['growth_rate']
model.fit(X,y)
print('Observations: R-squared = %0.2f' % model.score(X,y))
```

```python
X = df[df['type']=='AM'][['temp_24h_avg', 'rh_24h_avg', 'ws_24h_avg', 'lw_24h_avg', 'sw_24h_avg']]
y = df[df['type']=='AM']['growth_rate']
model.fit(X,y)
print('Morning observations: R-squared = %0.2f' % model.score(X,y))

X = df[df['type']=='PM'][['temp_24h_avg', 'rh_24h_avg', 'ws_24h_avg','lw_24h_avg', 'sw_24h_avg']]
y = df[df['type']=='PM']['growth_rate']
model.fit(X,y)
print('Evening observations: R-squared = %0.2f' % model.score(X,y))
```


Rather poor performance in both cases -- let's add additional features created from the same time series

```python
df['temp_24h_min'] = df['time'].apply(lambda x: dfb[(dfb.index<=x) & (dfb.index>=(x-timedelta(hours=24)))]['temp'].min())
df['temp_24h_max'] = df['time'].apply(lambda x: dfb[(dfb.index<=x) & (dfb.index>=(x-timedelta(hours=24)))]['temp'].max())
df['temp_24h_std'] = df['time'].apply(lambda x: dfb[(dfb.index<=x) & (dfb.index>=(x-timedelta(hours=24)))]['temp'].std())

df['rh_24h_min'] = df['time'].apply(lambda x: dfb[(dfb.index<=x) & (dfb.index>=(x-timedelta(hours=24)))]['rh'].min())
df['rh_24h_max'] = df['time'].apply(lambda x: dfb[(dfb.index<=x) & (dfb.index>=(x-timedelta(hours=24)))]['rh'].max())
df['rh_24h_std'] = df['time'].apply(lambda x: dfb[(dfb.index<=x) & (dfb.index>=(x-timedelta(hours=24)))]['rh'].std())

df['ws_24h_min'] = df['time'].apply(lambda x: dfb[(dfb.index<=x) & (dfb.index>=(x-timedelta(hours=24)))]['ws'].min())
df['ws_24h_max'] = df['time'].apply(lambda x: dfb[(dfb.index<=x) & (dfb.index>=(x-timedelta(hours=24)))]['ws'].max())
df['ws_24h_std'] = df['time'].apply(lambda x: dfb[(dfb.index<=x) & (dfb.index>=(x-timedelta(hours=24)))]['ws'].std())

df['lw_24h_min'] = df['time'].apply(lambda x: dfb[(dfb.index<=x) & (dfb.index>=(x-timedelta(hours=24)))]['lw'].min())
df['lw_24h_max'] = df['time'].apply(lambda x: dfb[(dfb.index<=x) & (dfb.index>=(x-timedelta(hours=24)))]['lw'].max())
df['lw_24h_std'] = df['time'].apply(lambda x: dfb[(dfb.index<=x) & (dfb.index>=(x-timedelta(hours=24)))]['lw'].std())

df['sw_24h_min'] = df['time'].apply(lambda x: dfb[(dfb.index<=x) & (dfb.index>=(x-timedelta(hours=24)))]['sw'].min())
df['sw_24h_max'] = df['time'].apply(lambda x: dfb[(dfb.index<=x) & (dfb.index>=(x-timedelta(hours=24)))]['sw'].max())
df['sw_24h_std'] = df['time'].apply(lambda x: dfb[(dfb.index<=x) & (dfb.index>=(x-timedelta(hours=24)))]['sw'].std())
```

```python
model_am = ElasticNet()
model_pm = ElasticNet()
X_am = df[df['type']=='AM'][[col for col in df.columns if (col.startswith('temp')
                                                           or col.startswith('rh')
                                                           or col.startswith('ws')
                                                           or col.startswith('lw')
                                                           or col.startswith('sw')
                                                          )]]
y_am = df[df['type']=='AM']['growth_rate']
model_am.fit(X_am,y_am)
print('Morning observations: R-squared = %0.2f' % model_am.score(X_am,y_am))

X_pm = df[df['type']=='PM'][[col for col in df.columns if (col.startswith('temp')
                                                            or col.startswith('rh')
                                                            or col.startswith('ws')
                                                           or col.startswith('lw')
                                                           or col.startswith('sw')
                                                           )]]
y_pm = df[df['type']=='PM']['growth_rate']
model_pm.fit(X_pm,y_pm)
print('Evening observations: R-squared = %0.2f' % model_pm.score(X_pm,y_pm))
```

```python
fig=plt.figure(figsize=(16,8))

# getting unzipped lists of feature names and corresponding coefficients, sorted by absolute decreasing coefficient
features_am, weights_am = zip(*sorted(zip(X_am.columns, model_am.coef_), key=lambda x: abs(x[1]), reverse=True))
features_pm, weights_pm = zip(*sorted(zip(X_pm.columns, model_pm.coef_), key=lambda x: abs(x[1]), reverse=True))

ax=fig.add_subplot(1,2,1)
plt.bar(range(len(weights_am)), weights_am)
ax.set_ylabel('Coefficient in linear regression')
plt.xticks([x+0.5 for x in range(len(weights_am))], features_am, rotation='vertical')
ax.set_title('Night ($R^2=%0.2f$)' % model_am.score(X_am,y_am), {'fontsize': 14, 'fontweight' : 'bold'})

ax=fig.add_subplot(1,2,2)
plt.bar(range(len(weights_pm)), weights_pm)
ax.set_ylabel('Coefficient in linear regression')
plt.xticks([x+0.5 for x in range(len(weights_pm))], features_pm, rotation='vertical')
ax.set_title('Day ($R^2=%0.2f$)' % model_pm.score(X_pm,y_pm), {'fontsize': 14, 'fontweight' : 'bold'}) 

plt.tight_layout()
plt.show()
```

### Restricting analysis to periods of fountain operation

```python
df=df[df['water']==1]
print('%d morning measurements' % len(df[df['type']=='AM']))
print('%d evening measurements' % len(df[df['type']=='PM']))
```

```python
df['temp_24h_min'] = df['time'].apply(lambda x: dfb[(dfb.index<=x) & (dfb.index>=(x-timedelta(hours=24)))]['temp'].min())
df['temp_24h_max'] = df['time'].apply(lambda x: dfb[(dfb.index<=x) & (dfb.index>=(x-timedelta(hours=24)))]['temp'].max())
df['temp_24h_std'] = df['time'].apply(lambda x: dfb[(dfb.index<=x) & (dfb.index>=(x-timedelta(hours=24)))]['temp'].std())

df['rh_24h_min'] = df['time'].apply(lambda x: dfb[(dfb.index<=x) & (dfb.index>=(x-timedelta(hours=24)))]['rh'].min())
df['rh_24h_max'] = df['time'].apply(lambda x: dfb[(dfb.index<=x) & (dfb.index>=(x-timedelta(hours=24)))]['rh'].max())
df['rh_24h_std'] = df['time'].apply(lambda x: dfb[(dfb.index<=x) & (dfb.index>=(x-timedelta(hours=24)))]['rh'].std())

df['ws_24h_min'] = df['time'].apply(lambda x: dfb[(dfb.index<=x) & (dfb.index>=(x-timedelta(hours=24)))]['ws'].min())
df['ws_24h_max'] = df['time'].apply(lambda x: dfb[(dfb.index<=x) & (dfb.index>=(x-timedelta(hours=24)))]['ws'].max())
df['ws_24h_std'] = df['time'].apply(lambda x: dfb[(dfb.index<=x) & (dfb.index>=(x-timedelta(hours=24)))]['ws'].std())

df['lw_24h_min'] = df['time'].apply(lambda x: dfb[(dfb.index<=x) & (dfb.index>=(x-timedelta(hours=24)))]['lw'].min())
df['lw_24h_max'] = df['time'].apply(lambda x: dfb[(dfb.index<=x) & (dfb.index>=(x-timedelta(hours=24)))]['lw'].max())
df['lw_24h_std'] = df['time'].apply(lambda x: dfb[(dfb.index<=x) & (dfb.index>=(x-timedelta(hours=24)))]['lw'].std())

df['sw_24h_min'] = df['time'].apply(lambda x: dfb[(dfb.index<=x) & (dfb.index>=(x-timedelta(hours=24)))]['sw'].min())
df['sw_24h_max'] = df['time'].apply(lambda x: dfb[(dfb.index<=x) & (dfb.index>=(x-timedelta(hours=24)))]['sw'].max())
df['sw_24h_std'] = df['time'].apply(lambda x: dfb[(dfb.index<=x) & (dfb.index>=(x-timedelta(hours=24)))]['sw'].std())
```

```python
model_am = ElasticNet()
X_am = df[df['type']=='AM'][[col for col in df.columns if (col.startswith('temp')
                                                           or col.startswith('rh')
                                                           or col.startswith('ws')
                                                           or col.startswith('lw')
                                                           or col.startswith('sw')
                                                          )]]
y_am = df[df['type']=='AM']['growth_rate']
model_am.fit(X_am,y_am)
print('All observations: R-squared = %0.2f' % model_am.score(X_am,y_am))
X_pm = df[df['type']=='PM'][[col for col in df.columns if (col.startswith('temp')
                                                            or col.startswith('rh')
                                                            or col.startswith('ws')
                                                           or col.startswith('lw')
                                                           or col.startswith('sw')
                                                           )]]
y_pm = df[df['type']=='PM']['growth_rate']
model_pm.fit(X_pm,y_pm)
print('Evening observations: R-squared = %0.2f' % model_pm.score(X_pm,y_pm))
```

```python
fig=plt.figure(figsize=(16,8))

# getting unzipped lists of feature names and corresponding coefficients, sorted by absolute decreasing coefficient
features_am, weights_am = zip(*sorted(zip(X_am.columns, model_am.coef_), key=lambda x: abs(x[1]), reverse=True))

ax=fig.add_subplot(1,2,1)
plt.bar(range(len(weights_am)), weights_am)
ax.set_ylabel('Coefficient in linear regression')
plt.xticks([x+0.5 for x in range(len(weights_am))], features_am, rotation='vertical')
ax.set_title('Night ($R^2=%0.2f$)' % model_am.score(X_am,y_am), {'fontsize': 14, 'fontweight' : 'bold'})

ax=fig.add_subplot(1,2,2)
plt.bar(range(len(weights_pm)), weights_pm)
ax.set_ylabel('Coefficient in linear regression')
plt.xticks([x+0.5 for x in range(len(weights_pm))], features_pm, rotation='vertical')
ax.set_title('Day ($R^2=%0.2f$)' % model_pm.score(X_pm,y_pm), {'fontsize': 14, 'fontweight' : 'bold'}) 

plt.tight_layout()
plt.show()
```

Growth Rate seems to be highly dependent on Relative humidity at night time. This suggests that high relative humidity in the previous 24 hours could cause dew formation on the Ice Stupa surface at night which could release additional latent heat and cause warming and lead to melting.
